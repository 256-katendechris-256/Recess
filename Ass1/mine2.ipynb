```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_one=pd.read_csv('mine.csv')
data_one
```

```python
data_one.head()
```

```python
#visualization
import matplotlib.pyplot as plt
plt.scatter(data_one['Pulse'],data_one['Maxpulse'],c=data_one['Duration'],s=data_one['Calories'])
plt.xlabel('Pulse')
# naming the x axis
plt.ylabel('Maximum Pulse')
# naming the y axis
plt.title('Scatter Graph')
plt.colorbar()
plt.show()
```

Line Graph

```python
data_two=pd.read_csv('mine.csv')

plt.plot(data_two['Pulse'])
plt.plot(data_two['Maxpulse'])

plt.xlabel('Pulse')
plt.ylabel('Maximum Pulse')

plt.title('Line Graph')
plt.show()
```

Bar Graph

```python
dat3=pd.read_csv('mine.csv')

plt.bar(dat3['Maxpulse'],dat3['Pulse'])
# plotting the bar graph of Max pulse vs Pulse for all activities in mine dataset
plt.xlabel('Max')
plt.ylabel('Pulse')

plt.title("Bar Graph")
plt.show()
```

Using seaborn

```python
dat4=pd.read_csv('mine.csv')
import seaborn as sns
#seaborn can be used together with matplotlib
sns.lineplot(x='Pulse',y='Maxpulse',data=dat4)

plt.title("Line Graph2")
plt.show()
```

```python
sns.scatterplot(x='Pulse',y='Maxpulse',
                hue='Calories',data=dat4)

plt.title("Scatter Graph2")
plt.show()
```

Barplot

```python
dat5=pd.read_csv('mine.csv')
sns.barplot(x='Pulse',y='Maxpulse',
                hue='Calories',data=dat5)

plt.title("Scatter Graph2")
plt.show()
```

Pie chart

```python
#creating a pie chart
dat6=pd.read_csv('mine.csv')

category_counts = dat6['Calories'].value_counts()

plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Calories')
plt.axis('equal')  # Ensures a circular pie chart
plt.show()



```
