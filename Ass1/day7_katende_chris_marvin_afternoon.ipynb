```python
sec=pd.read_csv('Work.csv')
sec
```

```python
#check for unique data
for i in sec.columns:
    print(i,"------",sec[i].unique())
```

```python
# Convert 'py-score' column to numeric format, replacing non-numeric values with NaN
sec['py-score'] = pd.to_numeric(sec['py-score'], errors='coerce')

# Remove rows with NaN values in 'py-score' column
sec1 = sec.dropna(subset=['py-score'])
sec1
```

```python
# Convert 'age' column to numeric format and handle missing values
sec['age'] = pd.to_numeric(sec['age'], errors='coerce')
```

```python
sec
```

```python
#dealing with duplicates
seq=sec.duplicated()
seq
#there are no duplicates
```

```python
# Calculate correlation coefficient between 'age' and 'py-score'
#am using sec1 coz its already dealt with
correlation = sec1['age'].corr(sec1['py-score'])
print("Correlation coefficient:", correlation)

#age increases as py,score decreases
```
